```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("salary(US$)")
plt.scatter(df["year"], df["per capita income (US$)"], color="red", marker=".")

```

```python
model = linear_model.LinearRegression()
model.fit(df[['year']],df['per capita income (US$)'])
```

```python
model.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("salary(US$)")
plt.scatter(df["year"], df["per capita income (US$)"], color="red", marker=".")
plt.plot(df.year, model.predict(df[['year']]))
```

```python
d = pd.read_csv("budget_vs_sales.csv")
d
```

```python
%matplotlib inline
plt.xlabel("budget in k US$")
plt.ylabel("Sales in k US$")
plt.scatter(d["budget_k$"],d["sales_k$"],color="blue",marker="*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[["budget_k$"]],d["sales_k$"])
```

```python
reg.predict([[26]])
```

```python
%matplotlib inline
plt.xlabel("budget in k US$")
plt.ylabel("Sales in k US$")
plt.scatter(d["budget_k$"],d["sales_k$"],color="blue",marker="*")
plt.plot(d["budget_k$"],reg.predict(d[["budget_k$"]]))
```
