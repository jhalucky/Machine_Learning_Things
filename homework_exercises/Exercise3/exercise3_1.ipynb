```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def predict_using_sklearn():
    reg = LinearRegression()
    reg.fit(df[['math']],df['cs'])
    return reg.coef_, reg.intercept_

predict_using_sklearn()
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 50000
    n = len(x)
    learning_rate = 0.0001

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n)*sum(val**2 for val in y-y_predicted)
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        if i % 100 == 0:
            print(f"m={m_curr}, b={b_curr}, iteration={i}")

    return m_curr, b_curr

        # return m_curr, b_curr

if __name__ == "__main__":
    df = pd.read_csv("test_scores.csv")
    x = np.array(df.math)
    y = np.array(df.cs)

    m,b = gradient_descent(x,y)
    print("Using gradient descent m{} and intercept{}".format(m,b))

    m_sklearn, b_sklearn = predict_using_sklearn()
    print("Using sklearn m{} and intercept{}".format(m_sklearn, b_sklearn))

```

```python
model = LinearRegression()
model.fit(x.reshape(-1,1), y)

print("Sklearn: m =", model.coef_[0], ", b =", model.intercept_)
```

```python
y_pred_sklearn = m_sklearn*x + b_sklearn

%matplotlib inline
plt.xlabel("math score")
plt.ylabel("cs score")
plt.scatter(x, y_pred_sklearn, color="blue", marker="*")
plt.plot(x,y_pred_sklearn,linestyle="--",label="sklearn linear regression")
plt.title("comparison of best fit line")
```
